Viết hàm tìm kiếm nhị phân
trên một danh sách đã
được sắp xếp tăng dần.
Trả về chỉ số của phần tử
hoặc -1 nếu không tìm
thấy. mẫu hàm: def
binary_search(arr, left,
right, target):


```python
def kt(a,k):
    l=0
    r=len(a)-1
    mid =(l+r)//2
    while l<=r:
        if (a[mid]>k):
            r=mid-1
            mid = (l+r)//2
        elif (a[mid]<k):
            l=mid+1
            mid=(l+r)//2
        else :
            return mid 
if __name__=='__main__':
    a = list(map(int,input().split()))
    k = int(input())
    print(kt(a,k))
```

W7A4
Viết hàm tìm phần tử có số
lần xuất hiện nhiều nhất
trong danh sách. Nếu có
nhiều phần tử cùng số lần,
trả về phần tử xuất hiện
trước.


```python
def kt(n):
    ln=0
    for i in range(len(n)):
        k=n.count(n[i])
        if k>ln:
            ln=k
            t=n[i]
    return t
n =list(map(int,input().split()))
print(kt(n))
```

W7A5
Cho danh sách các số
nguyên và một số X, đếm
tất cả các cặp (a, b) sao
cho a + b = X. c


```python
a = list(map(int, input().split()))
x = int(input())
dem = 0
for i in range(len(a)):
    for j in range(i + 1, len(a)):
        for k in range(len(a)):
            if k != i and k != j:
                if a[i] + a[j] == x * a[k]:
                    dem += 1
print(dem)
        
```

W7A7
Cho một danh sách các số
nguyên nums, hãy tìm độ
dài của dãy con tăng
nghiêm ngặt dài nhất
(không nhất thiết phải liền kề)

```python
import ast
a= ast.literal_eval(input())
dp=[1]*len(a)
for i in range(len(a)):
    for j in range(i):
        if a[j]<a[i]:
            dp[i]=max(dp[i],dp[j]+1)
print((dp))
```

danh sách các khoảng thời

độ dài nhỏ nhất của
khoảng thời gian [li, ri] sao
cho li ≤ q ≤ ri.
Nếu không tồn tại khoảng
thời gian nào như vậy, hãy
trả về -1 cho truy vấn đó.
solve in < O(m.n), hint:
binary search in the
queries list


```python
import ast 
a= ast.literal_eval(input())
b = ast.literal_eval(input())
c=[]
for i in b :
    ln=99999
    for j in a:
        if j[0]<=i<=j[1]:
            l=j[1]-j[0]+1
            if ln>l:
                ln=l
    if ln == 99999:
        c.append(-1)
    else :
        c.append(ln)
print(c)
                

```

W7A9
Cho một danh sách các số
nums, với mỗi phần tử
nums[i], hãy đếm xem có
bao nhiêu phần tử bên phải
nhỏ hơn nums[i].
Trả về danh sách các số
đếm.
Efficient solution: using
heap or binary search tree

```python
import ast 
a = ast.literal_eval(input())
d=[0]*len(a)
for i in range(len(a)):
    for j in range(i+1,len(a)):
        if a[i]>a[j]:
            d[i]+=1
print(d)
```
