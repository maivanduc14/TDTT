Viết hàm kiểm tra một số có phải số chính phương hay không? Hàm trả về kiểu bool, có tham số là một số

```python
from math import*
def kt(n):
    if n<2:
        return False
    else :
        for i in range(2,isqrt(n)+1):
            if n%i==0:
                return False
    return True
if __name__=='__main__':
    n =int(input())
    print(kt(n))
```

W5A18 VIết hàm kiểm tra một số có phải số đối xứng không, xử lý trên dữ liệu kiểu số nguyên.


```python
def kt(n):
    a=[]
    while (n>0):
        a.append(n%10)
        n=n//10
    l=len(a)//2
    for i in range(l):
        if a[i]!=a[len(a)-1-i]:
            return False
    return True
if __name__=='__main__':
    n = int(input())
    print(kt(n))
```

W5A19 VIết hàm kiểm tra một số có phải số kỳ diệu hay không, xử lý trên dữ liệu kiểu số nguyên

```python
from math import*
def kt(n):
    tong=1
    for i in range(2,isqrt(n)+1):
        if n%i==0:
            tong =tong+i
            if i*i!=n:
                tong=tong+n//i
    if tong==n:
        return True
    return False
if __name__=='__main__':
    n=int(input())
    print(kt(n))
```

W5A20 Viết hàm detect_number_type(s) trả về 'int' nếu s là số nguyên, 'float' nếu là số thực, ngược lại 'neither'.


```python
def detect_number_type(s):
    try:
        int(s)
        return "int"
    except:
        try:
            float(s)
            return "float"
        except:
            return "neither"
if __name__=='__main__':
    n=input()
    print(detect_number_type(n))
```

W5A23 Viết hàm quadratic_roots_count(a, b, c) trả về số nghiệm thực của phương trình a*x^2 + b*x + c = 0.

```python
def tn(a,b,c):
    denta=b**2-4*a*c
    if denta<0:
        x='Vô Nghiệm'
        return x
    elif denta==0:
        x= -b/(2*a)
        return x
    else :
        x1= (-b+denta**0.5)/(2*a)
        x2= (-b-denta**0.5)/(2*a)
        return x1,x2
if __name__=='__main__':
    a,b,c = map(float,input().split())
    print(*tn(a,b,c))
```

W5A24 Viết hàm are_collinear(x1,y1,x2,y2,x3,y3) kiểm tra ba điểm có thẳng hàng hay không.

```python
def kt(x1, y1, x2, y2, x3, y3):
    if (x2 - x1) * (y3 - y1) == (x3 - x1) * (y2 - y1) :
        return True
    return False
if __name__=='__main__':
    x1, y1, x2, y2, x3, y3= map(int,input().split())
    print(kt(x1, y1, x2, y2, x3, y3))
```

W5A28 Viết hàm tính góc ∠ABC (độ) từ tọa độ A(x1,y1), B(x2,y2), C(x3,y3).


```python
from math import*
def goc(x1,y1,x2,y2,x3,y3):
    BA=((x2-x1)**2+(y2-y1)**2)**0.5
    BC=((x3-x2)**2+(y3-y2)**2)**0.5
    t=((x1-x2)*(x3-x2)+(y1-y2)*(y3-y2))/(BA*BC)
    return acos(t)
if __name__=='__main__':
    x1,y1,x2,y2,x3,y3 = map(int,input().split())
    print(goc(x1,y1,x2,y2,x3,y3))
    
```

```python
def kt(n,k):
    if n<k :
        return False
    else :
        x=1 ; y=1 ; z=1
        for i in range(1,(n-k+1)):
            x=x*i
        for i in range(1,k+1):
            y=y*i
        for i in range(1,n+1):
            z=z*i
        tich= z//(x*y)
        return tich
if __name__=='__main__':
    n, k = map(int,input().split())
    print(kt(n,k))
```
