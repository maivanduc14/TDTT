Nhập vào một số nguyên đại diện cho tháng trong năm (nhuận).
In ra số ngày trong tháng đó 

```python
n = int(input())
match n:
    case 1|3|5|7|8|10|12:
        print(f"Tháng {n} có 31 ngày")
    case 4|6|9|11:
        print (f"Tháng {n} có 30 ngày ")
    case 2:
        print (f"Tháng {n} có 29 ngày ")
    case _:
        print (f"Thang {n} không có trong năm")
```

```python
n = int(input())
if (n==1 or n==3 or n==5 or n==7 or n==8 or n==10 or n==12 ):
    print (f"Thang {n} có 31 ngày ")
elif (n==4 or n==6 or n==9 or n==11):
    print(f"Tháng {n} có 30 ngày ")
elif (n==2):
    print(f"Tháng {n} có 29 ngày ")
else :
    print (f"Tháng {n} có không có trong năm ")
```
