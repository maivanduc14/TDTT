W4A18 Nhập vào số nguyên dương a và b, in toàn bộ ước chung của a và b 

```python
a , b = map(int,input().split())
from math import*
g=gcd(a,b)
for i in range(1,isqrt(g)+1):
    if g%i==0 :
        if i*i==g:
            print(i,end=' ')
        else :
            print(i,g//i,end=' ')
```

W4A20 Nhập vào một số nguyên dương n, kiểm tra xem n có phải là số dạng 2^k hay không 


```python
def kt(n):
    n=bin(n)[2:]
    for i in range(1,len(n)):
        if n[i]!='0':
            return False
    return True
n=int(input())
print(kt(n))
```

W4A25
Nhập vào một dãy số nguyên, ngưng nhập khi người dùng nhập -1.
Sau khi nhập xong, in số lớn nhất, số nhỏ nhất trong những số vừa nhập

```python
ln=-9999999
nn= 9999999
while(True):
    n = int(input())
    if n==-1:
        break
    else :
        ln=max(ln,n)
        nn=min(nn,n)
print(ln,nn)
```

Dãy số fibonacci là dãy số được định nghĩa như sau: 1, 1, 2, 3, 5, 8, 13,... với số kế tiếp sẽ
bằng tổng hai số trước đó
Nhập vào số nguyên dương A, hãy tìm số trong dãy số fibonacci lớn nhất nhưng không vượt
quá A 

```python
A = int (input())
b=0
c=1
while (b+c<=A):
    tong=b+c
    b,c=c,tong
print (c)
```

W4A29

W4A31
Nhập vào một chuỗi, hãy đếm xem trong chuỗi có bao nhiêu ký tự in hoa, bao nhiêu ký tự in thường , bao nhiêu kí tự chữ số 

```python
n=input()
dem1=0
dem2=0
dem3=0
for i in n :
    if i.isupper():
        dem1+=1
    elif i.islower():
        dem2+=1
    else:
        dem3+=1
print(dem1,dem2,dem3)
```
