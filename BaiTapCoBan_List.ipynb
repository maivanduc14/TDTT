### Bài 3: Tìm giá trị trung bình của 1 list các số nguyên dương

```python
def tong(a):
    tong=0
    dem=0
    for i in a :
        if i>0:
            tong+=i
            dem+=1
    if dem!=0:
        return round((tong/dem),2)
    else :
        return 'invalid'
if __name__=='__main__':
    a=list(map(int,input().split()))
    print(tong(a))
```

### Bài 4: Kiểm tra một list số nguyên có phải là đối xứng hay không

```python
def kt(a):
    l=(len(a)//2)
    for i in range(l):
        if a[i]!=a[len(a)-1-i]:
            return False
    return True
if __name__=='__main__':
    a = list(map(int,input().split()))
    print(kt(a))
```

### Bài 5: Đưa tất cả các số bằng 0 trong list về đầu của list. Các số khác không thay đổi thứ tự trước sau
3 4 6 7 1 9 0 0 23 0 0 34 67 89
0 0 0 0 3 4 6 7 1 9 23 34 67 89

```python
a = list(map(int,input().split()))
l = len(a)
b=[]
for i in a :
    if i !=0:
        b.append(i)
c=[0]*(l-len(b))+b
print(*c)
```

### Bài 6: Xóa tất cả các số giống nhau trong list, Các số khác giữ nguyên vị trí trước sau

### Bài 7: Kiểm tra một list số nguyên có phải là dãy tăng, dãy giảm hay vô danh
1 3 5 6 7 8 9: Tang
9 8 7 6 5 4 0: Giam
1 2 3 7 1 0 2: Vo danh
1 2 2 4 5 7: Vo danh

```python
a = list(map(int,input().split()))
b = sorted(a)
c= list(a)
c.sort(reverse=True)
if (a==b):
    print('Tang')
elif (a==c):
    print('Giam')
else :
    print('Vo danh')
```

### Bài 8: Kiểm tra một list số nguyên có phải là ziczac khong
1 3 2 5 2 8 7 9 2 100: ziczac
34 2 56 7 90 20 30 7 120: ziczac

```python
a = list(map(int,input().split()))
def kt(a):
    for i in range(1,len(a)-1):
        if a[i]-a[i-1]>0 and a[i+1]-a[i]<0:
            continue
        elif a[i]-a[i-1]<0 and a[i+1]-a[i]>0:
            continue
        else :
            return False
    return True
print(kt(a))
```

### Bài 9: Tìm và in ra tất cả các vị trí của số nguyên k trong một danh sách số nguyên cho trước. Nếu không có k trong ds thì in ra "-1".
VD: k = 3, [1,3,4,4,3,3,4,3,6,1] => k xuất hiện tại các vị trí (trả về 1 list) [1, 4, 5, 7]
k = 10, [1,3,4,4,3,3,4,3,6,1] => k không xuất hiện trong ds => trả về "-1"

```python
k = int(input())
a =list(map(int,input().split()))
c=[]
for i,x in enumerate(a):
    if x==k:
        c.append(i)
if (len(c)!=0):
    print(c)
else :
    print(-1)
```
